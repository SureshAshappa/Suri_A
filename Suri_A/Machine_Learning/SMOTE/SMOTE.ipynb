```python
import pandas as pd
import sklearn
from sklearn.utils import resample
```

```python
data = pd.read_csv('Downloads/UpSampDemo.csv')
```

```python
data = pd.DataFrame(data)
```

```python
data
```

```python
 # 21% are catg 1, imbalanced data! 3/14 # 21% are 1
```

```python
data.default.value_counts()
```

```python
# separate minority and majority classes
```

```python
not_default = data[data.default==0] #11
not_default
```

```python
len(not_default) #11
```

```python
default = data[data.default==1] # 3
default
```

```python
len(default) # 3
```

```python
#__________________________________ upsample minority_with replacement
from sklearn.utils import resample
default_upsampled1 = resample(default,
                          replace=True, # sample with replacement
                          n_samples=len(not_default), # match number in majority class, ie 11
                          random_state=27) # reproducible results
```

```python
# combine majority and upsampled minority
```

```python
default_upsampled1
```

```python
upsampled080321 = pd.concat([not_default, default_upsampled1]) 
upsampled080321
```

```python
# check new class counts
upsampled080321.default.value_counts() #11 11, You should be Happy!
```

```python
upsampled080321.to_csv('upsampled140321.csv')
```

```python
#This is a type of data augmentation for the minority class 
#and is referred to as the Synthetic Minority Oversampling 
#Technique, or SMOTE for short. In this tutorial, you will discover
#the SMOTE for oversampling imbalanced classification datasets.
```
